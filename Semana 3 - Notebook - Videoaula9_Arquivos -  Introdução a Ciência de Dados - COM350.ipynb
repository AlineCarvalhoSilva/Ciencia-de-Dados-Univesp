#Abrir arquivo (vamos olhar o conteúdo dele, é uma informação de texto copiada da Wikipedia - https://pt.wikipedia.org/wiki/Machado_de_Assis)

##Vamos subir esse arquivo para o Colab (lembrando que toda vez que ele parar a execução o arquivo será excluído)

```python
arquivo=open('machado_de_assis.txt','r')
```

#Ler linhas e fechar arquivo e mostrar conteúdo


```python
conteudo = arquivo.readlines()
arquivo.close()
```

```python
print(conteudo)
print('-----')
for linha in conteudo:
    print(linha)
```

#Alterar em todo o texto a palavra "Rio de Janeiro" por "Ribeirão Preto"

```python
x=''
for linha in conteudo:
    if linha!='':
        x+=linha

print(x)
x=x.replace('Rio de Janeiro','Ribeirão Preto')
print('--------')
print(x)
```

#Gravar (Gerar um novo arquivo)

```python
arquivo_novo = open('machado_assis_2.txt', 'w')
arquivo_novo.writelines(x)
arquivo_novo.close()
```

#Trabalhando com arquivos *CSV*

###Abrindo Arquivo CSV - Cotação das ações da Tesla

###https://www.kaggle.com/rpaguirre/tesla-stock-price?select=Tesla.csv+-+Tesla.csv.csv (quase 7 anos de arquivo)

```python
import csv
with open('kaggle_tesla.csv','r') as f:
    leitor=csv.reader(f, delimiter=',')
    for linha in leitor:
        data=linha[0]
        fechamento=linha[4]
        volume=linha[5]
        print('Data: ',data,'--Fechamento: ',fechamento,'--Volume: ',volume )
```
